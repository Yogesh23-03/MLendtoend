#### BIVARIATE ANALYSIS ( Is gender has any impact on student's performance ? ) 

```python
gender_group = df.groupby('gender').mean(numeric_only=True)
gender_group
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Step 1: Group by gender and take mean of numeric columns
gender_group = df.groupby('gender').mean(numeric_only=True)

print(gender_group)  # optional (to see values)

# Step 2: Define labels (what we are comparing)
X = ['Average Score', 'Math Score']

# Step 3: Extract values using .loc (IMPORTANT)
female_scores = [
    gender_group.loc['female', 'average'],
    gender_group.loc['female', 'math_score']
]

male_scores = [
    gender_group.loc['male', 'average'],
    gender_group.loc['male', 'math_score']
]

# Step 4: Bar positions
x = np.arange(len(X))   # [0,1]
width = 0.35

# Step 5: Plot
plt.figure(figsize=(10,6))

plt.bar(x - width/2, female_scores, width, label='Female', color='pink')
plt.bar(x + width/2, male_scores, width, label='Male', color='blue')

# Step 6: Labels & Title
plt.xlabel('Score Type')
plt.ylabel('Scores')
plt.title('Gender vs Scores Comparison')
plt.xticks(x, X)
plt.legend()

# Step 7: Show values on bars
for i, v in enumerate(female_scores):
    plt.text(i - width/2, v + 1, f"{v:.1f}", ha='center')

for i, v in enumerate(male_scores):
    plt.text(i + width/2, v + 1, f"{v:.1f}", ha='center')

plt.show()
```

#### Insights 
- On an average females have a better overall score than men.
- whereas males have scored higher in Maths.
